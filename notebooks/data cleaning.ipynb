```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.describe(include = 'all') # to check summary statistics of numerical column but use all to show all
```

```python
#check missing values
df.isnull().sum() #we see that review rating has 37 nulls, what do we do about that, fill with mean/median
```

```python
 #we saw review rating has 37 nulls, what do we do about that, replace null with mean/median
#we genarally would choose median over mean beacuse mean can eb affected by outliers, and not median, but we can't fill all rating with median of 
#entire column and move on, yes it can work but its not the best practice, beacuse rating can vary a lot based on category
#we will find a median rating for each acatgeory and replace each null by its real category median
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum() #check if we have any nulls now, and we don't
```

```python
# Renaming columns according to snake casing for better readability and documentation - otherwise you'd have to remember exactly how they are
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column for age group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#check if the 2 columns conatin the same info
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
